```python
# 将所有测试集中的图片大小尺寸变为 16 的倍数
from PIL import Image, ImageOps
import os

def resize_to_multiple_of_sixteen(image_path, output_path):
    # 打开图片
    with Image.open(image_path) as img:
        width, height = img.size
        # 计算新的尺寸，保证宽度和高度都是16的倍数
        new_width = ((width + 15) // 16) * 16
        new_height = ((height + 15) // 16) * 16
        
        if new_width != width or new_height != height:
            # 如果尺寸需要改变，则进行调整
            resized_img = img.resize((new_width, new_height), resample=Image.Resampling.LANCZOS)
        else:
            # 尺寸已经是16的倍数，直接使用原图
            resized_img = img
        
        # 保存调整后的图片
        resized_img.save(output_path)

def process_images_in_directory(directory, output_dir):
    # 确保输出目录存在
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            file_path = os.path.join(directory, filename)
            out_file_path = os.path.join(output_dir, filename)
            resize_to_multiple_of_sixteen(file_path, out_file_path)
            print(f"Processed {filename}")

# 使用示例
input_directory = './data/snow100k/test/input1'  # 输入图片所在目录
output_directory = './data/snow100k/test/input'  # 输出图片存储目录
process_images_in_directory(input_directory, output_directory)
```
